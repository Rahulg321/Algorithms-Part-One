```python
"""_summary_
        
        Creating a doubly linked list using my way and understanding
        will try and implement all the reqd basic methods from my understanding and then
        check the correct way from the course
"""


class Node:
    def __init__(self, value, next = None, prev = None):
        self.value = value
        self.next = next
        self.prev = prev

class LinkedList():
    def __init__(self) -> None:
        self.head = None
        self.tail = None
    
    def forward_traversal(self):
    #check if the list is empty
        if self.head == None:
            return f'The list is forwardly empty'
        else:
            temp = self.head
            while temp != None:
                print(temp.value)
                temp = temp.next
    
    def backward_traversal(self):
        if self.head == None:
            return f'The list is backwardly empty'
        else:
            temp = self.tail
            while temp != None:
                print(temp.value)
                temp = temp.prev   
    
    def delete_entire_list(self):
        if self.head is None:
            return f'The list is empty'
        else:
            self.head = None 
            self.tail = None
    
    def delete_node(self, location):
        """
            function deletes a node from a doubly linked list
        """
        #if the list is empty
        if self.head is None:
            return f'The list is empty cant delete anything new'
        #if there is only one node
        elif self.head == self.tail:
            print('The list contained only one element')
            self.head = self.tail = None
            
        else:
            #delete from the beginning
            if location == 0 or location == 1:
                self.head.next.prev = None
                self.head = self.head.next
            #deleting the last Node
            elif location == -1:
                self.tail.prev.next = None
                self.tail = self.tail.prev
            
            #deleting node at a specific location
            else:
                temp = self.head
                index = 1
                
                while index != location:
                    temp = temp.next
                    index += 1
            
                temp.prev.next = temp.next
                temp.next.prev = temp.prev
                
    def search_in_list(self, value):
        if self.head is None:
            return f'The list is empty'
        else:
            temp = self.head
            if temp.value == value:
                return f'Element found at head'
            else:
                index = 1
                
                while temp != None:
                    if temp.value == value:
                        return f'Node with the value {value} found at index {index}'
                    temp = temp.next
                    index += 1
                
                return f'NODE WITH THE VALUE {value} DOESNT EXIST!!!!!'
    
    def insert_node(self, value, location):
        """
            This function expects the value of the node to be passed in
            and the location at which the new node is to be inserted
        """  

        if self.head is None:
            return f'The list is empty'
        else:
            new_node = Node(value)
            if location == 0 or location == 1:
            #insert at the beginning of the doubly list
                new_node.next = self.head
                self.head.prev = new_node
                self.head = new_node
            elif location == -1:
            #inserting at the end of the doubly list
                self.tail.next = new_node
                new_node.prev = self.tail
                self.tail = new_node 
            else:
                temp = self.head
                index = 1
                
                while index != location:
                    temp = temp.next
                    index += 1
                
                temp_prev = temp.prev
                new_node.next = temp
                new_node.prev = temp_prev
                temp.prev = new_node
                temp_prev.next = new_node
                
    

circular_list = LinkedList()

first = Node(1)
second = Node(2)
third = Node(3)

first.next = second
second.prev = first
second.next = third
third.prev = second

circular_list.head = first
circular_list.tail = third

# print('Inserting at the second position of the list')
# circular_list.insert_node('John Doe', 2)


print('Deleting the second Node')
circular_list.search_in_list(3343)
# print()
# print('Forward traversal')
# circular_list.forward_traversal()
# print()
# print('Backward traversal')
# circular_list.backward_traversal()


        
```
