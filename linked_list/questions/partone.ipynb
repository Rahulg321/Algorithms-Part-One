```python
#Creating a general Linked List Class for our studies
#yield is a highly efficient genrator function
import random


class Node:
    def __init__(self, value) -> None:
        self.value = value
        self.next = None
        self.prev = None
    
    def __str__(self):
        return (str(self.value))


class LinkedList():
    
    def __init__(self):
        self.head = None
        self.tail = None
    
#this function makes our class iterable
    def __iter__(self):
        temp = self.head
        
        while temp:
            yield temp.value
            temp = temp.next
    
    def __str__(self) -> str:
        values = [str(x) for x in self]
        return ' -> '.join(values)

    def __len__(self):
        result = 0
        temp = self.head
        
        while temp:
            result += 1
            temp = temp.next
        
        return result

#creating a default add method, only adds a node to the end of the list
    def add(self, value):
        node = Node(value)
        if self.head is None:
            self.head = node
            self.tail = node
        else:
            self.tail.next = node
            self.tail = node
        
        return self.tail

#this func generates a random list for us 
    def generate_list(self, n, min_val, max_val):
        self.head = self.tail = None
        
        for _ in range(n):
            self.add(random.randint(min_val, max_val))
        
        return self
        

customLL = LinkedList()
customLL.generate_list(10,0,10)
print(customLL)
print(len(customLL))
```
