```python
#insertion operations in SingleLinkedList
class Node:
    def __init__(self, value = None):
        self.value = value 
        self.next = None 
    
    
class linkedlist:
    def __init__(self):
        self.head = None
    
    def insert_at_beginning(self, value):
        new_node = Node(value)
        
        new_node.next = self.head
        self.head = new_node
        
    
    def insert_at_given(self, prev, value):
        """
        this function expects a previous node to be passed in as arguement
        """
        new_node = Node(value)
        
        new_node.next = prev.next
        prev.next = new_node
    
    def insert_at_end(self, value):
        new_node = Node(value)
        
        temp = self.head
        while temp.next != None:
            temp = temp.next
        
        temp.next = new_node
        
    
    def print_list(self):
        temp = self.head
        
        while temp != None:
            print(temp.value)
            temp = temp.next
    
        
first = Node(34)
second = Node(56)
first.next = second
single_linked_list = linkedlist()

single_linked_list.head = first



single_linked_list.insert_at_beginning(68)
single_linked_list.insert_at_end(99)

single_linked_list.insert_at_given(prev = first, value = 'inserted at a given location')

#printing the list
single_linked_list.print_list()

        

        
```

```python
#traversal of a single linked list

#insertion operations in SingleLinkedList
#by maintaining a tail attribute and using a single function
class Node:
    def __init__(self, value = None):
        self.value = value 
        self.next = None 
    
    
class linkedlist:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def insert_node(self, value, location = None):
        new_node = Node(value)
#check whether list is empty or not 
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            #insert at the beginning
            if location == 0:
                new_node.next = self.head
                self.head = new_node
            #insert at the end
            elif location == -1:
                self.tail.next = new_node
                self.tail = new_node
            else:
            #inserting at a given location
                temp = self.head
                index = 1
                while index != location - 1:
                    temp = temp.next
                    index += 1
                    
                
                print(f'Value of temp is {temp.value}')
                next_node = temp.next
                temp.next = new_node
                new_node.next = next_node
                    
            
    def traversal(self):
        
        temp = self.head
        #check if the list is empty
        if temp is None:
            return 'The single linked list does not exist'
        
        else:
            while temp != None:
                print(temp.value)
                temp = temp.next
        

single_linked_list = linkedlist()    
first = Node(1)
second = Node(2)
third = Node(3)
fourth = Node(4)


first.next = second
second.next = third
third.next = fourth

single_linked_list.head = first
single_linked_list.tail = fourth


#printing the list
single_linked_list.traversal()


#Time Complexity -> O(n) 
#Space Complexity -> O(1) creating only one temporary node

        
```
