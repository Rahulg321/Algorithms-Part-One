```python
#my way
#queue with fixed capacity
class Queue:
    def __init__(self) -> None:
        self.queue = []
    
    def enqueue(self, value):
        self.queue.append(value)
    
    def dequeue(self):
        if self.is_empty():
            return f'the list is empty'
        else:
            self.queue.pop(0)
            
    
    def peek(self):
        """
        returns the first element without removing
        """
        return self.queue[0]

    def delete_queue(self):
       self.queue = None
       return 
    
    def is_empty(self):
        if self.queue == []:
            return True
        return False
    
    def __len__(self):
        return len(self.queue)
    
    # def __iter__(self):
    #     for x in self.queue:
    #         yield str(x)
    
    def __str__(self) -> str:
        values = [str(x) for x in self.queue]
        return " ".join(values)


q1 = Queue()


q1.enqueue(1)
q1.enqueue(2)
q1.enqueue(3)

print(q1)


q1.delete_queue()


print(q1.is_empty())

```
