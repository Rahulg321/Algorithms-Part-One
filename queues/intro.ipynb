```python
#queue with fixed capacity

class Queue:
    def __init__(self, size) -> None:
        self.queue = [None] * size
        self.size = size
        self.start = -1
        self.top = -1
    

    def enqueue(self, value):
        
        if self.is_full():
            return f'the queue is currently full'
        
        else:
            if self.top + 1 == self.size:
                self.top = 0
            
            else:
                self.top += 1
                
                if self.start == -1:
                    self.start += 1
                    
            self.queue[self.top] = value
        
        return f'the element has been inserted'
            
        
    def dequeue(self):
        if self.is_empty():
            return f'the list is empty, plz insert an element'
        else:
            element = self.queue[self.start]
            self.queue[self.start] = None
            
            if self.start == self.top:
                self.start = -1
                self.top = -1
                
            elif self.start + 1 == self.size:
                self.start = 0 
                
            else:
                self.start += 1 
            
            
            return element
                
            
    
    def peek(self):
        """
        returns the first element without removing
        """
        if self.is_empty():
            return f'the list is empty, plz insert an element'
        
        return self.queue[self.start]

    def delete_queue(self):
       self.queue = [None] * self.size
       self.start = -1
       self.top = -1
    
    def is_empty(self):
        if self.top == -1:
            return True
        else:
            return False
    
    def is_full(self):
        if self.top == self.size - 1 and self.start == 0:
            return True

        elif self.top + 1 == self.start:
            return True 
        
        else:
            return False
    
    def __len__(self):
        return self.size
    
    # def __iter__(self):
    #     for x in self.queue:
    #         yield str(x)
    
    def __str__(self) -> str:
        values = [str(x) for x in self.queue]
        return " ".join(values)


q1 = Queue(5)


q1.enqueue(1)
q1.enqueue(2)
q1.enqueue(3)
q1.enqueue(4)
q1.enqueue(5)
q1.enqueue(6)
print(q1)

q1.dequeue()
q1.dequeue()

print(q1)


```

```python
#creating a queue using Linked List
class Node:
    def __init__(self, value) -> None:
        self.value = value
        self.next = None
        self.prev = None
    
    def __str__(self) -> str:
        return self.value

class LinkedList:
    def __init__(self) -> None:
        self.head = None
        self.tail = None
    
    def __iter__(self):
        temp = self.head
        while temp:
            yield temp.value
            temp = temp.next
    
    # def __str__(self) -> str:
    #     values = []
    #     temp = self.head
    #     while temp:
    #         values.append(temp.value)
    #         temp = temp.next
        
    #     return "  ".join(values)
        # values = [str(x) for x in self]
        # return " -> ".join(values)

class Queue:
    def __init__(self) -> None:
        self.customLL = LinkedList()
    
    def enqueue(self, value):
        node = Node(value)
        if self.customLL.tail is None:
            self.customLL.head = node
            self.customLL.tail = node
        else:
            self.customLL.tail.next = node
            self.customLL.tail = node
            
    def dequeue(self):
        
        if self.isEmpty():
            return f'The list is empty'
        
    #if there is only one node
        elif self.customLL.head == self.customLL.tail:
            element = self.customLL.head.value
            self.customLL.head = None
            self.customLL.tail = None
            return element
        
    #if there are multiple nodes   
        else:
            element = self.customLL.head.value
            self.customLL.head = self.customLL.head.next
            return element
    
    def peek(self):
        if self.isEmpty():
            return f'The list is empty'
        element = self.customLL.head.value
        return element

    def isEmpty(self):
        if self.customLL.head is None:
            return True
        else:
            return False
    
    def delete(self):
        self.customLL.head = None
        self.customLL.tail = None
            
    def __str__(self) -> str:
        values = [str(x) for x in self.customLL]
        return " ".join(values)
        

q1 = Queue()
q1.enqueue(1)
q1.enqueue(2)
q1.enqueue(3)
q1.enqueue(4)
print(q1)
q1.dequeue()
q1.dequeue()
print(q1)
    
```

```python
#implemented queue using the collections module
#collections module provides some built in containers and allows us to store elements and iterate over them
from collections import deque

custom_q = deque(maxlen=3)
custom_q.append(1)
custom_q.append(2)
custom_q.append(3)
print(custom_q)

#if we add an element and it exceeds the size
#then it automatically removes the first element to may way for that element
custom_q.append(50)
print(custom_q)

#returns and removes the last element
print(custom_q.popleft())
print(custom_q)

custom_q.clear()
print(custom_q)
```
